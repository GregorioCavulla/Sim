```python
import pandas as pd

url = "./Online-Retail-France.xlsx"
df = pd.read_excel(url)
```

```python
df.head()
```

```python
u_description = df['Description'].nunique()
print("Description has {} unique values".format(u_description))
```

```python
df['Description'] = df['Description'].str.strip()
new_u_description = df['Description'].nunique()
print("Description has {} unique values after cleaning".format(new_u_description))
```

```python
inno_null = df['InvoiceNo'].isnull().sum()
print("InvoiceNo has {} null row".format(inno_null))
df_cleaned = df.dropna(subset=['InvoiceNo']).copy()
```

```python
size_w_credit = df_cleaned.shape[0]
df_cleaned = df_cleaned[~df_cleaned['InvoiceNo'].str.startswith('C')]
size_wo_credit = df_cleaned.shape[0]
print("Dimension with credit {}, dimension without credit {}".format(size_w_credit, size_wo_credit))
```

```python
size_w_credit = df_cleaned.shape[0]
df_cleaned = df_cleaned[~df_cleaned['Description'].str.contains('POSTAGE')]
size_wo_credit = df_cleaned.shape[0]
print("Dimension with Postage {}, dimension without postage {}".format(size_w_credit, size_wo_credit))
```

```python
# 1. Raggruppa per InvoiceNo e Description, sommando la Quantity
# Questo gestisce il caso in cui lo stesso prodotto appaia più volte nello stesso scontrino
basket = df.groupby(['InvoiceNo', 'Description'])['Quantity'].sum()
# 2. Unstack: sposta 'Description' dalle righe alle colonne
# 3. fillna(0): sostituisce i NaN (prodotti non comprati in quella fattura) con 0
# 4. reset_index: riporta InvoiceNo come colonna normale invece di indice
basket = basket.unstack().fillna(0).reset_index()
basket.head()
```

```python
basket = basket.set_index('InvoiceNo')
basket = basket > 0
basket.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

step = 0.01
min_support = 1
min_rules = 20
metric = 'lift'
min_threshold = 1

while True:
    frequent_itemsets = apriori(basket, min_support=min_support, use_colnames=True)
    if frequent_itemsets.shape[0]>0:
        rules = association_rules(frequent_itemsets, metric=metric, min_threshold=min_threshold)
    if frequent_itemsets.shape[0] >0 and rules.shape[0]>=min_rules:
        break
    else:
        min_support -= step

rules
```

```python
rules = rules.sort_values(by=['lift', 'confidence'], ascending=False).reset_index()
rules.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(rules['lift'], rules['confidence'])
plt.show()
```
