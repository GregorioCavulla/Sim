# Imports, and setup

```python
import pandas as pd


url = "./Online-Retail-France.xlsx"
```

# 1. Upload, show size, and small portion

```python
df = pd.read_excel(url)
rows, cols = df.shape
print(f"Dataframe size: {rows} rows, {cols} columns")
df.head()
```

# 2. Print number of unique Description Values

```python
# 2. Print number of unique Description Values
n_descriptions = df['Description'].nunique()
print(f"Number of unique Description values: {n_descriptions}")
```

# 3. Clean Description column with str.strip()

```python
cleaned_descriptions = df['Description'].str.strip()
ncleaned_descriptions = cleaned_descriptions.nunique()
print(f"Number of unique Description values after cleaning: {ncleaned_descriptions}")
```

# 4. Remove rows with missing InvoiceNo

```python
df = df[df['InvoiceNo'].notna()]
rows, cols = df.shape
print(f"New dataframe size after removing rows with missing InvoiceNo: {rows} rows, {cols} columns")
nremoved_descriptions = df['Description'].nunique()
print(f"Number of unique Description values after removing rows with missing InvoiceNo: {nremoved_descriptions}")
```

# 5. Remove Credit trasactions (InvoiceNo starting with 'C')

```python
df = df[df['InvoiceNo'].str.startswith('C') == False]
rows, cols = df.shape
print(f"New dataframe size after removing credit transactions: {rows} rows, {cols} columns")
nremoved_descriptions = df['Description'].nunique()
print(f"Number of unique Description values after removing credit transactions: {nremoved_descriptions}")
```

# 6. remove item POSTAGE

```python
df = df[df['Description']!="POSTAGE"]
rows, cols = df.shape
print(f"New dataframe size after removing item POSTAGE: {rows} rows, {cols} columns")
nremoved_descriptions = df['Description'].nunique()
print(f"Number of unique Description values after removing item POSTAGE: {nremoved_descriptions}")
```

# 7.cnosilidate items into one trasnaction per row with onehotencoded Group by InvoceNo and Description sum on Quanity, using unstack function, reset index, fill missing with 0 store in basket

```python
# pls help me to complete the code
#After the cleanup, we need to consolidate the items into one trans-
#action per row with products one–hot–encoded. To do so, group by
#InvoiceNo and Description computing a sum on Quantity, use the
#Pandas unstack function to move the items from rows to columns, reset
#the index, fill the missing with zero, store the result in a new dataframe
#basket and inspect it
basket = df.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0)
basket = basket.set_index('InvoiceNo')
basket.head()
```

# 8. convert into boolean values

```python
def encode(x):
    if x <= 0:
        return False
    else:
        return True
basket = basket.applymap(encode)
basket.head()
```

# 9. find max value of min_support

```python
from mlxtend.frequent_patterns import apriori, association_rules

min_support = 1
rules = []
target_rule_count = 20

while min_support > 0:
    
    frequent_itemsets = apriori(
        basket,
        min_support=min_support,
        use_colnames=True
    )
    
    if not frequent_itemsets.empty:
        rules = association_rules(
            frequent_itemsets,
        )
        
    if len(rules) >= target_rule_count:
        break
    
    if len(rules) > 0:
        print(f"min_support={min_support:.2f} produced {len(rules)} rules.")
    
    min_support -= 0.01

print(f"Final result:\nmin_support={min_support:.2f} produced {len(rules)} rules.")
```

# 10. association rules using metric='lift' and min_threshold=1

```python
min_support = 1
rules = []
target_rule_count = 20

print("Using metric='lift' and min_threshold=1")

while min_support > 0:
    
    frequent_itemsets = apriori(
        basket,
        min_support=min_support,
        use_colnames=True
    )
    
    if not frequent_itemsets.empty:
        rules = association_rules(
            frequent_itemsets,
            metric='lift',
            min_threshold=1
        )
        
    if len(rules) >= target_rule_count:
        break
    
    if len(rules) > 0:
        print(f"min_support={min_support:.2f} produced {len(rules)} rules.")
    
    min_support -= 0.01

print(f"Final result:\nmin_support={min_support:.2f} produced {len(rules)} rules.")
```

# 11. Scatterplot rules sorted by descending lift and confidence

```python
rules.sort_values(by=['lift', 'confidence'], ascending=False, inplace=True)
rules.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(rules['lift'], rules['confidence'], alpha=0.5)
plt.title('Association Rules')
plt.xlabel('Lift')
plt.ylabel('Confidence')
plt.show()
```
